```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("housedata.csv")

# Check for missing values
df = df[['city', 'price']].dropna()

# Aggregating average price per city
average_prices = df.groupby('city').mean().sort_values(by='price', ascending=False)

# Limiting to top 20 cities for better visualization
top_n = 20
average_prices = average_prices.head(top_n)

# Plotting the bar graph
plt.figure(figsize=(12, 6))
plt.bar(average_prices.index, average_prices['price'], color='skyblue')
plt.xlabel('City')
plt.ylabel('Average House Price')
plt.title('Top 20 Cities by Average House Price')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
