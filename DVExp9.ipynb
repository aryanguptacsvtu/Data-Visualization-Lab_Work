```python
import matplotlib.pyplot as plt
import numpy as np

# Generate overlapping data
np.random.seed(0)
x = np.random.normal(5, 0.5, 200)
y = np.random.normal(5, 0.5, 200)

# Create side-by-side subplots
plt.figure(figsize=(12, 5))

# --- Plot 1: Without transparency ---
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', alpha=1.0)  # Fully opaque
plt.title("Without Partial Transparency")
plt.xlabel("X Axis(Random Data)")
plt.ylabel("Y Axis")
plt.grid(True)

# --- Plot 2: With partial transparency ---
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='blue', alpha=0.3)  # Partially transparent
plt.title("With Partial Transparency")
plt.xlabel("X Axis(Random Data)")
plt.ylabel("Y Axis")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create smaller clustered dataset
np.random.seed(1)
x = np.repeat([1, 2], 40)  # 20 points for each category = 40 total
y = np.random.normal(5, 0.3, 80)

# Apply jitter to x values
x_jittered = x + np.random.uniform(-0.1, 0.1, size=len(x))

# Create side-by-side subplots
plt.figure(figsize=(12, 5))

# --- Plot 1: Without jittering ---
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='green')
plt.title("Without Jittering")
plt.xlabel("Category (1 & 2)")
plt.ylabel("Y Axis")
plt.xticks([1, 2])
plt.grid(True)

# --- Plot 2: With jittering ---
plt.subplot(1, 2, 2)
plt.scatter(x_jittered, y, color='green')
plt.title("With Jittering")
plt.xlabel("Category (1 & 2)")
plt.ylabel("Y Axis")
plt.xticks([1, 2])
plt.grid(True)

plt.tight_layout()
plt.show()
```
