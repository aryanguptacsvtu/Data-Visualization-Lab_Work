```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Generate a right-skewed dataset (Exponential Distribution)
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000)

# Convert into a DataFrame
df = pd.DataFrame(data, columns=['Skewed_Data'])

# Calculate Skewness before transformation
original_skewness = skew(df['Skewed_Data'])
print(f"Original Skewness: {original_skewness:.4f}")

# Plot Original Distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Skewed_Data'], bins=30, kde=True)
plt.title("Original Skewed Data")

# Apply Log Transformation to reduce skewness
df['Transformed_Data'] = np.log1p(df['Skewed_Data'])  # log1p handles zero values safely

# Calculate Skewness after transformation
transformed_skewness = skew(df['Transformed_Data'])
print(f"Transformed Skewness: {transformed_skewness:.4f}")

# Plot Transformed Distribution
plt.subplot(1, 2, 2)
sns.histplot(df['Transformed_Data'], bins=30, kde=True)
plt.title("Transformed (Reduced Skewness) Data")

plt.show()
```
