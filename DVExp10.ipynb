```python
import matplotlib.pyplot as plt

# Enhanced Sample Data
x = [1, 2, 3, 4, 5]
y1 = [1, 3, 2, 5, 4]
y2 = [2, 2.5, 3, 3.5, 4]
y3 = [5, 4, 3, 2, 1]
y4 = [1, 4, 1, 4, 1]

# Create figure and axis
plt.figure(figsize=(10, 6))

# 1. Named color
plt.plot(x, y1, color='red', label='Named color (red)', linewidth=2, marker='o')

# 2. Hex color code
plt.plot(x, y2, color='#00FF00', label='Hex color (#00FF00)', linewidth=2, marker='s')

# 3. RGB tuple
plt.plot(x, y3, color=(0.1, 0.2, 0.5), label='RGB tuple (0.1, 0.2, 0.5)', linewidth=2, marker='^')

# 4. Grayscale
plt.plot(x, y4, color='0.5', label='Grayscale (0.5)', linewidth=2, marker='D')

# Title, legend, and grid
plt.title('Using Different Color Codes', fontsize=16)
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('X-axis(Random Data)')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```
